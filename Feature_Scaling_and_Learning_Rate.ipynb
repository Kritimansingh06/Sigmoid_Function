```python
import math, copy
import numpy as np
import matplotlib.pyplot as plt
```

```python
import numpy as np

x_train = np.array([
    [952, 2, 1, 65],
    [1244, 3, 2, 64],
    [1947, 3, 2, 17],
    [1720, 3, 2, 42],
    [1960, 4, 2, 15],
    [1310, 2, 1, 20],
    [864, 1, 1, 80],
    [2100, 4, 3, 10],
    [1450, 3, 2, 30],
    [1100, 2, 1, 50]
])

y_train = np.array([
    271.5,232,509,380,520,
    310,180,620,405,250
])

x_features = ['size(sqft)','bedrooms','floors','age']

# learned parameters
w = np.array([0.16595, -7.15833, 47.54096, -2.03163])
b = 143.0386
```

```python
def gradient(x,y,w,b):
    m = x.shape[0]
    dj_dw = 0
    dj_db = 0

    for i in range(m):
        f_w = np.dot(x[i],w)+b
        dj_dw_i = (f_w - y[i])*x[i]
        dj_db_i = (f_w - y[i])
        dj_dw = dj_dw + dj_dw_i
        dj_db = dj_db + dj_db_i

    dj_dw = dj_dw/m
    dj_db = dj_db/m

    return dj_dw , dj_db
```

```python
def cost_function(x,y,w,b):
    m = x.shape[0]
    cost = 0

    for i in range(m):
        f_w = np.dot(x[i],w)+b
        cost_t = (f_w - y[i])**2
        cost = cost + cost_t

    cost = cost/(2*m)

    return cost
```

```python
def Gradient_descent(x,y,w_init,b_init,ite,gradient,cost_function,alpha):
    w = copy.deepcopy(w_init);
    b = b_init;
    J_history = []

    for i in range(ite):
        dj_dw , dj_db = gradient(x,y,w,b)

        w = w - alpha*dj_dw
        b = b - alpha*dj_db

        if(i<10000):
            J_history.append(cost_function(x,y,w,b))

        if i% math.ceil(ite / 10) == 0:
            print(f"Iteration {i:4d}: Cost {J_history[-1]:8.2f}   ")

    return w , b , J_history
```

```python
fig,ax = plt.subplots(1,4,figsize=(12 ,3),sharey=True)
for i in range(len(ax)):
    ax[i].scatter(x_train[:,i],y_train)
    ax[i].set_xlabel(x_features[i])
ax[0].set_ylabel("Price in $1000's")
```

```python
alpha = 9.9e-7
w1,b1,J_hist = Gradient_descent(x_train,y_train,w,b,10000,gradient,cost_function,alpha)
print(f"Updated value of W: {w1} , Updated value of B: {b1}")
```

```python
def zscore_normalization(x):
    mu = np.mean(x,axis=0)
    sigma = np.std(x,axis=0)
    x_norm = (x-mu)/sigma

    return (mu,sigma,x_norm)
```

```python
mu = np.mean(x_train,axis=0)
sigma = np.std(x_train,axis=0)
x_mu = x_train-mu
x_norm = (x_train-mu)/sigma

fig,ax = plt.subplots(1,3,figsize=(12,3))

ax[0].scatter(x_train[:,0],x_train[:,3])
ax[0].set_ylabel(x_features[0])
ax[0].set_xlabel(x_features[3])
ax[0].set_title("Unnormalized")

ax[1].scatter(x_mu[:,0],x_mu[:,3])
ax[1].set_xlabel(x_features[3])
ax[1].set_title("X-mu")

ax[2].scatter(x_norm[:,0],x_norm[:,3])
ax[2].set_xlabel(x_features[3])
ax[2].set_title("X_Normalization")
```
