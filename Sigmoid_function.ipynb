```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Testing for exponential 
x = np.array([1,2,3,4,5])
y = np.exp(x)

print(f"Input Value: {x}")
print(f"Output Value: {y}")
```

```python
# Sigmoid Function
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
z_val = np.arange(-10,10)
y_val = sigmoid(z_val)

np.set_printoptions(precision=3) 
print("  Input       Output")
print(np.c_[z_val,y_val])
```

```python
# Plotting Sigmoid Function
plt.plot(z_val,y_val,c='b')
plt.xlabel("z")
plt.ylabel("Sigmoid(z)")
plt.title("Learning Sigmoid function")
plt.show()
```

```python
x_train = np.array([0., 1, 2, 3, 4, 5])
y_train = np.array([0,  0, 0, 1, 1, 1])

w_in = np.zeros((1))
b_in = 0.5
```

```python
f_w = w_in*x_train + b_in
```

```python
pos = y_train == 1
neg = y_train == 0

plt.scatter(x_train[pos],y_train[pos],marker='x',s=80,label='y=1')
plt.scatter(x_train[neg],y_train[neg],marker='o',s=100,label='y=0')
plt.plot(x_train,f_w)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
